```python
# Q2. Write a NumPy program to count frequency of unique values in a NumPy array.
```

```python
import numpy as np

arr = np.array([2, 3, 4, 2, 5, 4, 2, 6, 4, 3, 2, 5, 6, 3])

unique_values, counts = np.unique(arr, return_counts=True)

for value, count in zip(unique_values, counts):
    print(f"{value} occurs {count} time(s)")
```

```python
# Q5. Write a NumPy program to find the length, dimensions, size of a Numpy array.
```

```python
# Q13. Write a NumPy program to find the most frequent value in an array.
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 2, 2, 3, 3, 3])

most_frequent_value = np.bincount(arr).argmax()

print("Array:")
print(arr)
print("\nMost frequent value:", most_frequent_value)
```

```python
# Q15. Write a NumPy program to create a random array with 1000 elements and compute the average, variance, standard deviation of the array elements.
Sample output:
Average of the array elements: -0.0255137240796
Standard deviation of the array elements: 0.984398282476
Variance of the array elements:Â 0.969039978542
```

```python
import numpy as np

arr = np.random.randn(1000)

average = np.mean(arr)

variance = np.var(arr)

std_deviation = np.std(arr)

print("Average of the array elements:", average)
print("Standard deviation of the array elements:", std_deviation)
print("Variance of the array elements:", variance)
```

```python
# Q1. Write a Python program to draw a line plot for total profit of all months with proper label = Month number and label = Total Profit.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv')
print(df)

month_number = df['month_number']
profits = ['total_profit']

for profit in profits:
    total_profit = df[profits]
    plt.plot(month_number, total_profit, marker='o', label=profit)

plt.xlabel('Month number')
plt.ylabel('Total Profit')

plt.title('Total Profit by Month')

plt.show()

```

```python
# Q3. Write a Python program to draw a multiline plot for all product sales data.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv')
print(df)
month_number = df['month_number']
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

for product in products:
    sales_data = df[product]
    plt.plot(month_number, sales_data, marker='o', label=product)

plt.xlabel('Month number')
plt.ylabel('Product Sales')
plt.title('Product Sales by Month')

plt.legend()

plt.show()
```

```python
# Q8. Write a Python program to calculate total sale data in last year for each product and draw a Pie chartÂ forÂ theÂ same.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_data.csv')

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

sales_data = []
for product in products:
    total_sales = df[product].sum()
    sales_data.append(total_sales)

plt.pie(sales_data, labels=products, shadow=True, autopct="%1.1f%%")

plt.title("Total Sales in Last Year")
plt.show()
```
