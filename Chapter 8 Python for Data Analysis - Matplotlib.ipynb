## Matplotlib Labels and Title

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

plt.title('Sports Watch Data', fontdict = font1, loc = 'center')
plt.xlabel('Average Pulse', fontdict = font2)
plt.ylabel('Calorie Burnage', fontdict = font2)

plt.plot(x, y, marker = '*', ms = 10, color = 'r')
plt.show()
```

## Matplotlib Adding Grid Lines

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

font1 = {'family':'serif', 'color':'blue', 'size':25}
font2 = {'family':'serif', 'color':'red', 'size':15}

plt.title('Sports Watch Data', fontdict = font1, loc = 'center')
plt.xlabel('Average Pulse', fontdict = font2)
plt.ylabel('Calorie Burnage', fontdict = font2)

plt.plot(x, y, marker = '*', ms = 10, color = 'r')

plt.grid()
plt.grid(axis = 'x', color = 'blue', linestyle = ':', linewidth =1 )
plt.grid(axis = 'y', color = 'green', linestyle = ':', linewidth =1) 
#plt.grid(axis = 'y', color = 'green', linestyle = '--', linewidth =1) 

plt.show()
```

## Matplotlib Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np

fruits = np.array([25, 35, 25, 15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Dates']
plt.pie(fruits, labels = mylabels, startangle = 90, autopct = '%1.1f%%')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fruits = np.array([25, 35, 25, 15])
mylabels = ['Apples', 'Bananas', 'Cherries', 'Dates']
myexplode = [0.2, 0, 0.1, 0]

plt.pie(fruits, labels = mylabels, explode = myexplode, shadow = True, startangle = 90, autopct = '%1.1f%%')

plt.show()
```

```python
#There are 100 students in the class in which 25 students are failed, 25 are in a distinction, 20 in the first class, and 30 in the second class.

import matplotlib.pyplot as plt
import numpy as np

Students = np.array([30, 20, 25, 25])
mylabels = ['First Class', 'Second Class', 'Distinct', 'Failed']
myexplode = [0.1, 0, 0.1, 0]

plt.pie(Students, labels = mylabels, explode = myexplode, shadow = True, startangle = 90, autopct = '%1.1f%%')

plt.show()
```

```python
#There are 100 students in the class in which 25 students are failed, 25 are in a distinction, 20 in the first class, and 30 in the second class.

import matplotlib.pyplot as plt
import numpy as np

Students = np.array([30, 20, 25, 25])
mylabels = ['First Class', 'Second Class', 'Distinct', 'Failed']
mycolors = ['red', 'hotpink', 'b', '#4CAF50']
myexplode = [0.1, 0, 0.1, 0]

plt.pie(Students, labels = mylabels, colors = mycolors, explode = myexplode, shadow = True, startangle = 90, autopct = '%1.1f%%')

plt.show()
```

```python
#There are 100 students in the class in which 25 students are failed, 25 are in a distinction, 20 in the first class, and 30 in the second class.

import matplotlib.pyplot as plt
import numpy as np

Students = np.array([30, 20, 25, 25])
mylabels = ['First Class', 'Second Class', 'Distinct', 'Failed']
mycolors = ['red', 'hotpink', 'b', '#4CAF50']
myexplode = [0.1, 0, 0.1, 0]

plt.pie(Students, labels = mylabels, explode = myexplode, shadow = True, startangle = 90, autopct = '%1.1f%%')
plt.legend(title = 'Student Result', loc = 'center')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('sales_data.csv')
print(df)

profitlist = df['total_profit'].tolist()
monthlist = df['month_number'].tolist()

plt.xlabel('Month Number')
plt.ylabel('Profit in Rupees')
plt.title('Company Profit per Month')
plt.yticks(profitlist)
plt.xticks(monthlist)

plt.plot(monthlist, profitlist, marker = '*', color = 'r')
plt.show()
```
